## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
test1 = pd.read_csv('./data/input/metrics/visualization_metrics_kmeans_0.csv', index_col=0)
```

```python
test1 = test1.sort_values(by='k')
```

```python
test1.sort_values(by='combined_index')
```

```python
# Description of plot
plt.title('KMeans metrics of Question 1')
plt.xlabel('k ( with range [22,212] )')
plt.ylabel('Indicies')

# Plot the lines
x_range = test1['k']
plt.plot(x_range, -test1['db_scale'],label="Scaled DB index (Negated)", color='#b2df8a')
plt.plot(x_range, test1['ch_scale'], label="Scaled CH index", color='#1f78b4')
plt.plot(x_range, test1['sil'], label="Silhoutte Coefficient", color='#a6cee3')
plt.plot(x_range, test1['combined_index'], label="Combined", color="black")

# Horizontal line to show the max combined index
plt.axhline(0.105152, linestyle='dashed', color='grey')

# Adjusted settings for visual clarity
plt.ylim([-1.05,1.05])
plt.xlim([18,216])
plt.legend(bbox_to_anchor=(1,1), loc="upper left")

plt.show()
```
